# Fetch all race laps

```python
!pip install fastf1

import fastf1
import pandas as pd
import os

BASE = '/content/drive/MyDrive/tyre_project'
os.makedirs(os.path.join(BASE,'raw'), exist_ok=True)

fastf1.Cache.enable_cache('/content/fastf1_cache')

def fetch_race(year, gp_name, session_type='R'):
    session = fastf1.get_session(year, gp_name, session_type)
    session.load()
    laps = session.laps.copy()
    keep = ['Driver','Team','LapNumber','LapTime','Compound','IsPitLap','Stint','LapStartTime']
    for c in keep:
        if c not in laps.columns:
            laps[c] = None
    laps = laps[keep].reset_index(drop=True)
    laps['LapTime_s'] = laps['LapTime'].dt.total_seconds()
    laps['year'] = year
    laps['event'] = gp_name
    return laps

# Build schedule
year = 2024
schedule = fastf1.get_event_schedule(year)

# Remove testing
if 'EventFormat' in schedule.columns:
    schedule = schedule[schedule['EventFormat'].str.lower().ne('testing')]
else:
    schedule = schedule[~schedule['EventName'].str.contains(r'test|testing', case=False, na=False)]

frames = []
for _, row in schedule.iterrows():
    event_name = row['EventName']
    try:
        df_ev = fetch_race(year, event_name, 'R')
        df_ev['round'] = int(row['RoundNumber'])   # attach season order
        frames.append(df_ev)
        print(f"✓ {event_name}")
    except Exception as e:
        print(f"⚠️ Skipped {event_name}: {e}")

all_2024 = pd.concat(frames, ignore_index=True)

raw_path = os.path.join(BASE, 'raw', 'laps_2024.csv')
all_2024.to_csv(raw_path, index=False)
print(f"Saved RAW 2024 laps -> {raw_path} | rows={len(all_2024)}")
```

# Preprocess & clean data (remove pit/in/out laps, outliers, compute stints)

```python
import pandas as pd
import numpy as np
import os

RAW = os.path.join(BASE,'raw')
CLEAN = os.path.join(BASE,'clean')
os.makedirs(CLEAN, exist_ok=True)

df = pd.read_csv(os.path.join(RAW,'laps_2024.csv'))

# Normalize column names
df.rename(columns={'Driver':'driver','Team':'team','LapNumber':'lap_number',
                   'Compound':'compound','IsPitLap':'is_pit','Stint':'stint',
                   'LapStartTime':'lap_start_time','LapTime_s':'lap_time_s',
                   'Round':'round'}, inplace=True)

# Ensure round is integer
df['round'] = pd.to_numeric(df['round'], errors='coerce').astype('Int64')

# Boolean coercion
df['is_pit'] = df['is_pit'].astype(str).str.lower().isin(['true','1'])

# Filter
df = df[df['lap_time_s'].notna() & (df['lap_time_s'] > 0)].copy()
df['lap_number'] = pd.to_numeric(df['lap_number'], errors='coerce').fillna(0).astype(int)

# Compute stint_index
df = df.sort_values(['year','round','event','driver','lap_number']).reset_index(drop=True)
df['pit_marker'] = df['is_pit'].astype(int)
df['stint_index'] = 1 + df.groupby(['year','event','driver'])['pit_marker'].cumsum()

# Remove pit & out laps
grp = df.groupby(['year','event','driver'], sort=False, group_keys=False)
is_pit_next = grp['is_pit'].shift(-1).fillna(False)
df = df[~(df['is_pit'] | is_pit_next)].copy()

# Lap index in stint & stint length
df = df.sort_values(['year','round','event','stint_index','lap_number']).reset_index(drop=True)
df['lap_index_in_stint'] = df.groupby(['year','event','driver','stint_index']).cumcount()
df['stint_length'] = df.groupby(['year','event','driver','stint_index'])['lap_number'].transform('count')

# Remove short stints
df = df[df['stint_length'] >= 3].copy()

# Remove outliers
stats = df.groupby('driver')['lap_time_s'].agg(['mean','std']).reset_index()
df = df.merge(stats, on='driver', how='left')
df = df[df['lap_time_s'] <= df['mean'] + 4*df['std']].copy()
df.drop(columns=['mean','std','pit_marker'], inplace=True, errors='ignore')

# Fill compound
df['compound'] = df['compound'].fillna('Unknown').astype(str)

# --- Ordering ---
# Season order for events
ev_order = (df[['round','event']]
            .dropna()
            .drop_duplicates()
            .sort_values('round')['event']
            .tolist())
df['event'] = pd.Categorical(df['event'], categories=ev_order, ordered=True)

# Driver order by first appearance
first_seen = df.groupby('driver')['round'].min().sort_values(kind='stable')
drv_order = first_seen.index.tolist()
df['driver'] = pd.Categorical(df['driver'], categories=drv_order, ordered=True)

# Save
out_path = os.path.join(CLEAN, 'clean_laps_2024.csv')
df.to_csv(out_path, index=False)

print(f"Saved CLEAN 2024 laps -> {out_path} | rows={len(df)}")
print("Season order preview:", ev_order[:5])
print("Driver order preview:", drv_order[:5])
```

# Feature engineering + export a small sample

```python
import os, pandas as pd

CLEAN = os.path.join(BASE, 'clean')
SUB   = os.path.join(BASE, 'submission')
os.makedirs(SUB, exist_ok=True)

df = pd.read_csv(os.path.join(CLEAN, 'clean_laps_2024.csv'))

# Feature engineering: stint mean, relative, normalized
df['stint_mean']   = df.groupby(['year','event','driver','stint_index'])['lap_time_s'].transform('mean')
df['lap_time_rel'] = df['lap_time_s'] - df['stint_mean']
df['lap_time_norm']= df['lap_time_s'] / df['stint_mean']

# Save engineered full dataset (optional)
fe_path = os.path.join(CLEAN, 'clean_laps_2024_fe.csv')
df.to_csv(fe_path, index=False)

# Small sample for quick sharing
sample = df.head(1500)
sample_path = os.path.join(SUB, 'sample_stints_2024.csv')
sample.to_csv(sample_path, index=False)

print("Saved:", fe_path)
print("Saved sample:", sample_path, "| rows:", len(sample))
```

# EDA (figures saved to Drive, respects season order)

```python
import os, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FIG = os.path.join(BASE, 'figures_sample')
os.makedirs(FIG, exist_ok=True)

df = pd.read_csv(os.path.join(BASE, 'clean', 'clean_laps_2024_fe.csv'))

# ensure season order
ev_order = (df[['round','event']].dropna().drop_duplicates()
            .sort_values('round')['event'].tolist())
df['event'] = pd.Categorical(df['event'], categories=ev_order, ordered=True)

# Pick 5 sample races (adjust which ones you want)
sample_races = ['Bahrain Grand Prix', 'Monaco Grand Prix',
                'British Grand Prix', 'Italian Grand Prix', 'Abu Dhabi Grand Prix']

for race in sample_races:
    dfr = df[df['event']==race].copy()
    if dfr.empty:
        print(f"⚠️ No data for {race}")
        continue

    # 1. Lap time distribution
    plt.figure(figsize=(8,5))
    sns.histplot(dfr['lap_time_s'], bins=40)
    plt.xlabel('Lap time (s)')
    plt.title(f'{race} — Lap time distribution')
    plt.tight_layout()
    plt.savefig(os.path.join(FIG, f"{race}_hist_laptimes.png"), dpi=200)
    plt.close()

    # 2. Boxplot by compound
    plt.figure(figsize=(8,5))
    sns.boxplot(data=dfr, x='compound', y='lap_time_s')
    plt.title(f'{race} — Lap times by compound')
    plt.ylabel('Lap time (s)')
    plt.tight_layout()
    plt.savefig(os.path.join(FIG, f"{race}_box_compound.png"), dpi=200)
    plt.close()

    # 3. Degradation curves
    group = dfr.groupby(['compound','lap_index_in_stint'])['lap_time_s'].mean().reset_index()
    plt.figure(figsize=(9,5))
    for comp in group['compound'].unique():
        g = group[group['compound']==comp].sort_values('lap_index_in_stint')
        plt.plot(g['lap_index_in_stint'], g['lap_time_s'], label=comp)
    plt.xlabel('Lap index in stint')
    plt.ylabel('Average lap time (s)')
    plt.title(f'{race} — Degradation by compound')
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(FIG, f"{race}_degradation.png"), dpi=200)
    plt.close()

    print(f"✓ Plots saved for {race}")
```

```python
import numpy as np, pandas as pd, os
df = pd.read_csv(os.path.join(BASE, 'clean', 'clean_laps_2024_fe.csv'))

print('Rows (clean 2024):', len(df))
print('Unique drivers:', df['driver'].nunique())
print('Unique events:', df['event'].nunique())
print('Percent Unknown compound:', round(100*df['compound'].eq('Unknown').mean(), 2), '%')

# slope of degradation (s/lap) per compound
slopes = {}
for comp in df['compound'].unique():
    g = df[df['compound']==comp].groupby('lap_index_in_stint')['lap_time_s'].mean().reset_index()
    if len(g) >= 2:
        slopes[comp] = float(np.polyfit(g['lap_index_in_stint'], g['lap_time_s'], 1)[0])
print('Estimated slope (s per lap) by compound:', slopes)

# season order preview
order_preview = (df[['round','event']].drop_duplicates().sort_values('round').to_dict('records')[:8])
print('First 8 rounds:', order_preview)
```
